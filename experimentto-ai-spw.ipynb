```python
def preprocess_single_col(df) :
    df['No'] = np.repeat(np.arange(1, np.ceil(len(df) / 2).astype(int) + 1), 2)
    df_new = pd.DataFrame(columns = ['Pertanyaan', 'Jawaban'])
    for index in df['No'].unique() :
        tmp_data = df.loc[df.No == index, 'Pertanyaan dan jawaban']
        df_new.loc[index, ['Pertanyaan', 'Jawaban']] = [tmp_data.iloc[0], tmp_data.iloc[1]]
    return df_new
```

```python
def preprocess_unpad(excel_file, sheet_name = 'Sheet1') :
    data = pd.read_excel(excel_file, sheet_name = sheet_name)
    print(sheet_name)
    data['No'] = np.repeat(np.arange(1, np.ceil(len(data) / 2).astype(int) + 1), 2)
    data_new = pd.DataFrame(columns = ['Pertanyaan', 'Jawaban'])
    for index in data.No.unique() :
        tmp_data = data.loc[data.No == index, 'Pertanyaan dan jawaban']
        data_new.loc[index, ['Pertanyaan', 'Jawaban']] = [tmp_data.iloc[0], tmp_data.iloc[1]]
    data_new['Kategori'] = sheet_name
    return data_new
```

```python
from langchain.document_loaders import DataFrameLoader
from langchain import HuggingFaceHub, PromptTemplate, FAISS, HuggingFacePipeline
from langchain.vectorstores import Chroma
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain.embeddings import HuggingFaceEmbeddings, OpenAIEmbeddings, SpacyEmbeddings, HuggingFaceHubEmbeddings, CacheBackedEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.storage import InMemoryStore, LocalFileStore, RedisStore
import os
import openai

os.environ['HUGGINGFACEHUB_API_TOKEN'] = huggingface_key
os.environ['OPENAI_API_KEY'] = openai_key
```

```python
loader = DataFrameLoader(concated, page_content_column = 'Pertanyaan')
docs = loader.load()
```

```python
embeddings = OpenAIEmbeddings()
fs = LocalFileStore('./cache_embed/')
cache_embed = CacheBackedEmbeddings.from_bytes_store(
    embeddings, fs, namespace = embeddings.model
)
```

```python
vectordb = FAISS.from_documents(docs, cache_embed)
vectordb
```

```python
vectordb.similarity_search_with_score('Apa itu SPW?')
```

```python
vectordb.similarity_search_with_score('Bagaimana cara mengoperasikan mesin CNC?')
```

```python
vectordb.similarity_search_with_score('Apa Kegunaan dari STEM?')
```

```python
embedding_vector = cache_embed.embed_query('Apa itu SPW?')
docs_and_scores = vectordb.similarity_search_by_vector(embedding_vector)
docs_and_scores
```

```python
model = ChatOpenAI(model_name = 'gpt-4-1106-preview')
```

```python
retriever = RetrievalQA.from_chain_type(
    llm = model,
    chain_type = 'stuff',
    retriever = vectordb.as_retriever(search_kwargs = {'k' : 5, 'score_threshold' : 0.22}),
    return_source_documents = True
)
print(retriever)
```
